# Top 10 countries with highest and lowest populations

```python
df = df.sort_values('2022 Population', ascending = False)
df
```

```python
plt.figure(figsize=(10, 5))

bar = plt.bar(df['Country/Territory'][:10], df['2022 Population'][:10]/1000000)
plt.title('Most Populated Countries [2022]')
plt.xlabel('Countries')
plt.ylabel('Population (in millions)')
plt.xticks(rotation = 'vertical')
plt.bar_label(bar)
plt.show()

plt.figure(figsize=(10, 5))

bar = plt.bar(df['Country/Territory'][-10:], df['2022 Population'][-10:])
plt.title('Least Populated Countries [2022]')
plt.xlabel('Countries')
plt.ylabel('Population')
plt.xticks(rotation = 'vertical')
plt.bar_label(bar)
plt.show()
```

# Top 10 countries with highest and lowest population density

```python
df = df.sort_values('Density (per km²)', ascending = False)

plt.figure(figsize=(10, 5))

plt.bar(df['Country/Territory'][:10], df['Density (per km²)'][:10])
plt.title('Countries with highest population Density')
plt.xlabel('Countries')
plt.ylabel('Density (per km²)')
plt.xticks(rotation = 90)
plt.show()

plt.figure(figsize=(10, 5))

plt.bar(df['Country/Territory'][-10:], df['Density (per km²)'][-10:])
plt.title('Countries with lowest population Density')
plt.xlabel('Countries')
plt.ylabel('Density (per km²)')
plt.xticks(rotation = 90)
plt.show()
```

# Top 10 countries with highest and lowest growth rate

```python
df = df.sort_values('Growth Rate', ascending = False)

plt.figure(figsize=(10, 5))

plt.bar(df['Country/Territory'][:10], df['Growth Rate'][:10])
plt.title('Countries with highest Growth Rate')
plt.xlabel('Countries')
plt.ylabel('Growth Rate')
plt.ylim(1, 1.1)
plt.xticks(rotation = 90)
plt.show()

df = df.sort_values('Growth Rate', ascending = False)

plt.figure(figsize=(10, 5))

plt.bar(df['Country/Territory'][-10:], df['Growth Rate'][-10:])
plt.title('Countries with lowest Growth Rate')
plt.xlabel('Countries')
plt.ylabel('Growth Rate')
plt.ylim(0.9, 1)
plt.xticks(rotation = 90)
plt.show()
```

# Content-wise

```python
asia = df[df['Continent'] == 'Asia']
na = df[df['Continent'] == 'North America']
sa = df[df['Continent'] == 'South America']
africa = df[df['Continent'] == 'Africa']
europe = df[df['Continent'] == 'Europe']
oceania = df[df['Continent'] == 'Oceania']
```

```python
asia.head(3)
```

# Content-wise population

```python
continents = ['Asia', 'Europe', 'Africa', 'South America', 'North America', 'Oceania']
population_2022 = np.array([asia['2022 Population'].sum(),
                  europe['2022 Population'].sum(),
                  africa['2022 Population'].sum(),
                  sa['2022 Population'].sum(),
                  na['2022 Population'].sum(),
                  oceania['2022 Population'].sum()])

plt.figure(figsize=(10, 5))
plt.bar(continents, population_2022/1000000)
plt.title('Content-wise Population [2022]')
plt.xlabel('Continents')
plt.ylabel('Population (in millions)')
plt.xticks(rotation = 90)
plt.show()
```

```python
plt.figure(figsize=(7, 7))
plt.pie(population_2022, labels = continents, autopct='%0.1f%%')
plt.title('Continent-wise population ratio [2022]')
plt.show()
```

# Growth

```python
def growth(continent):
    return continent.sum()[2:10].values[::-1]
```

```python
asia_growth = growth(asia)/1000000
africa_growth = growth(africa)/1000000
na_growth = growth(na)/1000000
sa_growth = growth(sa)/1000000
europe_growth = growth(europe)/1000000
oceania_growth = growth(oceania)/1000000

years = ['1970', '1980', '1990', '2000', '2010', '2015', '2020', '2022']
```

```python
plt.figure(figsize=(10,7))

plt.plot(years, asia_growth, label ='Asia', marker = 'o')
plt.plot(years, sa_growth, label = 'South America', marker = 'o')
plt.plot(years, na_growth, label = 'North America', marker = 'o')
plt.plot(years, africa_growth, label = 'Africa', marker = 'o')
plt.plot(years, oceania_growth, label = 'Oceania', marker = 'o')
plt.plot(years, europe_growth, label = 'Europe', marker = 'o')

plt.title('Continent wise population growth')
plt.xlabel('Years')
plt.ylabel('Population (in millions)')
plt.legend()
plt.show()
```

# World population growth

```python
world_growth = growth(df)
world_growth
```

```python
plt.figure(figsize=(10, 7))

plt.plot(years, world_growth/1000000, marker = 'o')
plt.title('World population growth')
plt.xlabel('Years')
plt.ylabel('Population (in millions)')
plt.show()
```
